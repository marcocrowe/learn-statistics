## Q4.1.1. The relative frequency of Dice 2 = 4 , display your answer to 4 decimal places.



Each dice has 6 sides, so the probability of rolling a 6 is $\frac{1}{6}$. The probability of rolling a 6 on two dice is $\frac{1}{6} * \frac{1}{6} = \frac{1}{36}$.  
Dice 2 is 4 Dice 1 and Dice 2 are (1, 4) (2, 4) (3, 4) (4, 4) (5, 4) (6, 4).  
Sum is of possible outcome is 6  
Total possible outcomes is 36  
$\frac{6}{36} = \frac{1}{6} = 0.1667$

## Q4.1.2 The joint probability of Dice #1 = 4 and Dice #2 = 3. display your answer to 4 decimal places.

Each dice has a total of 6 sides, The total number of possible outcomes is 6 for one dice.  S= 6  

$ P(Dice1=4 \cap Dice2=3) $  
$ P(Dice1=4) = \frac{1}{6} $  
$P(\text{Dice2}=3) = \frac{1}{6}$  
$ P(Dice1=4 \cap Dice2=3) = P(Dice1=4) * P(Dice2=3) = \frac{1}{6} * \frac{1}{6} = \frac{1}{36} = 0.0278$

## Q4.1.3 The probability of Dice 1 been an even number given that Dice 2 = 6

The probability of Dice 1 been an even number given that Dice 2 = 6. display your answer to 4 decimal places.  
Count of Even number outcome on a dice is $\frac{3}{6} = \frac{1}{2} = 0.5$

$P(Dice1=Even | Dice2=6) = \frac{P(Dice1=Even \cap Dice2=6)}{P(Dice2=6)}$  
$P(Dice1=Even | Dice2=6) = \frac{P(Dice1=Even \cap Dice2=6)}{P(Dice2=6)} = \frac{3}{36} / \frac{6}{36} = \frac{3}{36} * \frac{36}{6} = \frac{3}{6} = 0.5$

### Q4.2 Bayes Theorem is a mathematical formula for calculating conditional probabilities. Define Bayes Theorem, use both the definition and formula to clarify your answer.



Bayes' theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event.

*For example:* if I throw 2 dice and I know that one of the dice is 4, we can use Bayes' theorem to calculate the probability of the other dice being 3.  The total count of possible outcomes is 36. Knowing the value of dice 2 is 4, the possible outcomes are (1, 4) (2, 4) (3, 4) (4, 4) (5, 4) (6, 4). The total count of possible outcomes is 6. The probability of dice 1 being 3 given that dice 2 is 4 is $\frac{1}{6}$.
$P(A|B) = \frac{P(B|A) * P(A)}{P(B)}$


Relative Frequency is the ratio (fraction or proportion) of the number of times an event occurs to the total number of trials or observations.

There axioms of probability theory are:

1. The probability of an event is a number between 0 and 1.
2. The probability of the sample space is 1.
3. Complementary probability: The probability of an event not occurring is 1 minus the probability of the event occurring. $P('A) = 1 - P(A)$
4.  The probability of the union of two mutually exclusive events is the sum of the probabilities of the individual events. The probability of the union of two events is the sum of the probabilities of the individual events minus the probability of the intersection of the two events. $P(A \cup B) = P(A) + P(B) - P(A \cap B)$
5.  Conditional probability: The probability of event A given that event B has occurred is the probability of the intersection of A and B divided by the probability of B. $P(A|B) = \frac{P(A \cap B)}{P(B)}$
